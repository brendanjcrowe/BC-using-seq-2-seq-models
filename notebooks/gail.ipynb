```python
from stable_baselines import GAIL
from stable_baselines.gail import ExpertDataset
```

```python
dataset = ExpertDataset(expert_path='expert_mountain_car.npz', traj_limitation=100, verbose=0)
model = GAIL('MlpPolicy', 'MountainCar-v0', dataset, verbose=1)
model.pretrain(dataset, n_epochs=10000, learning_rate=0.001)
model.learn(total_timesteps=100000)
```

```python
env = gym.make('MountainCar-v0')
for i in range(20)
obs = env.reset()
done = False
while not done:
    action, _states = model.predict(obs)
    obs, rewards, done, info = env.step(action)
    env.render()
env.close()
```
